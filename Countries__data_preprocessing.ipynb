````python
rename_col = {'Name': 'Country Name', 'GENC': 'Country Alpha', 'Area in Square Kilometers': 'Total Area', 'Population': 'Total_country_population','Life Expectancy at Birth, Both Sexes': 'Life Expectancy at Birth', 'Infant Mortality Rate, Both Sexes': 'Infant Mortality Rate' }
data_df.rename(columns=rename_col, inplace=True)
data_df
````

### Swapping columns in a DF
##### The total area moves to the third position while the total population occupies the last index

````python
cols = data_df.columns.tolist()
cols[-1], cols[3] = cols[3], cols[-1]
df_swapped = country_data[cols]
df_swapped
````

````python
df_swapped.drop(columns='Rate of Natural Increase', axis=1, inplace=True)
````

````python
len(df_swapped.columns)
df_swapped
````

````python
df_swapped.drop(columns='Life Expectancy at Birth, Males', axis=1, inplace=True)
````

````python
df_swapped['Net Migration Rate'] = pd.to_numeric(df_swapped['Net Migration Rate'], errors='coerce')
df_swapped.dtypes
````

````python
dd = df_swapped.columns.tolist()
dd[11]
````

#### A function that retrieves the Migration status

````python
def mig_status(df):
  column = 'Net Migration Rate'
  df['Migration'] = df[column] >= 0
  return df
new_data = mig_status(df_swapped)
````

````python
new_data.isna().sum()
````

````python
new_data[new_data['Country Alpha'].isna()]
````

#### Fill ```Namibia``` with the necessary iso name "NA"

````python
new_data['Country Alpha'] = new_data['Country Alpha'].fillna("NA")
````

````python
new_data.isna().sum()
````

````python
new_data[new_data['Net Migration Rate'].isna()]
````

````python
new_data = new_data.dropna()
preprocessed_data = new_data.copy()
````

````python
new_data.head()
````

````python
# preprocessed_data.iloc[300: 500]
new_data[new_data['Country Name'] == 'Gaza Strip']
new_data.iloc[400: 450]
````

````python
# Replacing incorrect country names
new_data['Country Name'] = new_data['Country Name'].replace('Gambia, The', 'Gambia')
new_data['Country Name'] = new_data['Country Name'].replace('Gaza Strip', 'State of Palestine')
# Replacing country alpha for State of palestine
new_data['Country Alpha'] = new_data['Country Alpha'].replace('XG', 'SP')
````
